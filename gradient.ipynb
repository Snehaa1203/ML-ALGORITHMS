```python
data=np.loadtxt(r"C:\Users\Hp\Downloads\train - train.csv",delimiter=',')
test=np.loadtxt(r"C:\Users\Hp\Downloads\test - test.csv",delimiter=',')
```

```python
def step_gradient(X, Y, m, learning_rate):
    m_slope=np.zeros(len(X[0]))
    for i in range(len(X)):
        for j in range(len(X[i])):
            m_slope[j]+=(-2/len(X))*(Y[i]-sum(m*X[i]))*X[i][j]
    new_m=m-(learning_rate*m_slope)
    return new_m

def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost+=(1/len(x)*(y[i]-sum(m*x[i]))**2)
    print(cost)

def gd(x, y, learning_rate, iterations):
    m=np.zeros(len(x[0]))
    for i in range(iterations):
        m=step_gradient(x, y, m, learning_rate)
        print("itr= ", i, "cost=", end=' ')
        cost(m, x, y)
    return m

def gradient_descent(x, y):
    iterations=200
    learning_rate=0.000000001
    x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m=gd(x, y, learning_rate, iterations)
    return m

x=data[:, :-1]
y=data[:, -1]
```
